```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
HM1f=le.fit_transform(HM1)
HM2f=le.fit_transform(HM2)
HM3f=le.fit_transform(HM3)
AM1f=le.fit_transform(AM1)
AM2f=le.fit_transform(AM2)
AM3f=le.fit_transform(AM3)
```

```python
HM1_f = HM1f.reshape(-1,1)
HM2_f = HM2f.reshape(-1,1)
HM3_f = HM3f.reshape(-1,1)
AM1_f = AM1f.reshape(-1,1)
AM2_f = AM2f.reshape(-1,1)
AM3_f = AM3f.reshape(-1,1)
```

```python
y=le.fit_transform(y)
y
```

```python
data2.groupby('FTR')['FTR'].agg('count')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
HM1_f1 = ohe.fit_transform(HM1_f)
HM2_f1 = ohe.fit_transform(HM2_f)
HM3_f1 = ohe.fit_transform(HM3_f)
AM1_f1 = ohe.fit_transform(AM1_f)
AM2_f1 = ohe.fit_transform(AM2_f)
AM3_f1 = ohe.fit_transform(AM3_f)
```

```python
type(HM1_f1)
```

```python
HM1_f2= HM1_f1.toarray()
HM2_f2= HM2_f1.toarray()
HM3_f2= HM3_f1.toarray()
AM1_f2= AM1_f1.toarray()
AM2_f2= AM2_f1.toarray()
AM3_f2= AM3_f1.toarray()
```

```python
type(HM1_f2)
```

```python
HM1_f2
```

```python
x_f=data2[['HTP','ATP','HTGD','ATGD','DiffFormPts']]
```

```python
import numpy
```

```python
x_final=numpy.hstack((x_f,HM1_f2,HM2_f2,HM3_f2,AM1_f2,AM2_f2,AM3_f2))
```

```python
type(y)
```

```python
x_final
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_final,y,test_size=0.3,random_state=0)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(random_state=42)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
y_pred == ytest
```

```python
model_accuracy=numpy.mean(y_pred  ==  ytest)
```

```python
model_accuracy
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(ytest, y_pred)
```

```python
sns.heatmap(cm,annot=True,fmt='d')
```
